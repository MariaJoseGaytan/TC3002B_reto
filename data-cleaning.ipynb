```python
import tokenize
import io
import os
```

```python
def clean_and_tokenize_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        source_code = f.read()
        
    token_stream = tokenize.generate_tokens(io.StringIO(source_code).readline)
    clean_tokens = []
    
    for tok in token_stream:
        if tok.type == tokenize.COMMENT:
            continue
        if tok.type == tokenize.STRING and tok.string.startswith(('"""', "'''")):
            continue

        clean_tokens.append(tok)
        
    return clean_tokens
```

```python
def tokens_to_clean_code(tokens):
    return tokenize.untokenize(tokens)
```

```python
def process_and_save_dataset():
    
    all_files = os.listdir("Data/original-codes")
    total_files = len(all_files)
    count = 0
    
    for index, file_name in enumerate(all_files, start=1):
        input_file_path = os.path.join("Data/original-codes/",file_name)
        output_file_path = os.path.join("Data/cleaned-code/",file_name)

        print(f"{index}/{total_files} Cleaning: {file_name}")

        tokens = clean_and_tokenize_file(input_file_path)
        clean_source = tokens_to_clean_code(tokens)

        if clean_source is not None:
            with open(output_file_path, 'w', encoding='utf-8') as out_file:
                out_file.write(clean_source)
            print("Saved")
            count+=1
        else:
            print("Failed")
    print(f"{count} files successfully cleaned")

    
```

```python
if __name__ == "__main__":
    process_and_save_dataset()
```
